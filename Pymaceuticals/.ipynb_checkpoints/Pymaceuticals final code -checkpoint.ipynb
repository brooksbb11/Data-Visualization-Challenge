```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df=pd.merge(mouse_metadata, study_results, on="Mouse ID")
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice. 249
combined_df["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_df = combined_df[combined_df.duplicated(["Mouse ID","Timepoint"])]
dup_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_df = combined_df[combined_df.duplicated(["Mouse ID"])]
dup_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_df = combined_df[combined_df["Mouse ID"]!="g989"]
combined_df
```

```python
# Checking the number of mice in the clean DataFrame.248
len(combined_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for ea
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
reg_df = combined_df.groupby("Drug Regimen")
t_mean = reg_df["Tumor Volume (mm3)"].mean()
t_median = reg_df["Tumor Volume (mm3)"].median()
t_var = reg_df["Tumor Volume (mm3)"].var()
t_stdev = reg_df["Tumor Volume (mm3)"].std()
t_SEM = reg_df["Tumor Volume (mm3)"].sem()
```

```python
# Assemble the resulting series into a single summary DataFrame.
results_df = pd.DataFrame([t_mean,t_median,t_var,t_stdev,t_SEM]).T
results_df.columns = ["Tumor Volume Mean",
                      "Tumor Volume Median",
                      "Tumor Volume Variance",
                      "Tumor Volume Std. Dev.",
                      "Tumor Volume Std. Err."]
results_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen
aggregations = {"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]}
# Using the aggregation method, produce the same summary statistics in a single line.
combined_df.groupby("Drug Regimen").agg(aggregations)
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_tp = combined_df.groupby("Mouse ID")
last_tp = last_tp["Timepoint"].max()
last_tp_df = pd.DataFrame(last_tp)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
t_volume_df = pd.merge(last_tp_df, combined_df, on = ["Mouse ID","Timepoint"])
t_volume_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
t_vol_capomulin = []
t_vol_ramicane = []
t_vol_infubinol = []
t_vol_ceftamin = []
# Create empty list to fill with tumor vol data (for plotting)
for i in range(len(t_volume_df)-1):
    if t_volume_df.loc[i, "Drug Regimen"] == "Capomulin":
        t_vol_capomulin.append(t_volume_df.loc[i, "Tumor Volume (mm3)"])
    if t_volume_df.loc[i, "Drug Regimen"] == "Ramicane":
        t_vol_ramicane.append(t_volume_df.loc[i, "Tumor Volume (mm3)"])
    if t_volume_df.loc[i, "Drug Regimen"] == "Infubinol":
        t_vol_infubinol.append(t_volume_df.loc[i, "Tumor Volume (mm3)"])
    if t_volume_df.loc[i, "Drug Regimen"] == "Ceftamin":
        t_vol_ceftamin.append(t_volume_df.loc[i, "Tumor Volume (mm3)"])
t_vol = [t_vol_capomulin, t_vol_ramicane, t_vol_infubinol, t_vol_ceftamin]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
outlier_list = []
for i in range(len(t_vol)):
    quartiles = pd.Series(t_vol[i]).quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f"\n{treatments[i]}:")
    print(f"The lower quartile of tumor volume is: {lowerq}.")
    print(f"The upper quartile of tumor volume is: {upperq}.")
    print(f"The interquartile range of tumor volume is: {iqr}.")
    print(f"The median of tumor volume is: {quartiles[0.5]}.")
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    for j in range(len(t_vol[i])):
        if (t_vol[i][j] < lower_bound) or (t_vol[i][j] > upper_bound):
            outlier = t_vol[i][j]
            outlier_list.append(outlier)
            print(f"The outliers are {outlier_list}.")    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
distribution = dict(markerfacecolor = "r", marker = "D")
fig1, ax1 = plt.subplots()
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(t_vol, flierprops = distribution)
ax1.set_title("Final Tumor Volume across Four Regimens", fontsize = 14)
ax1.set_xticklabels(treatments)
plt.show()
```
